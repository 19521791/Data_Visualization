```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('california_cities.csv')
cities.head()
```

```python
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
```

```python
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8, 6))
plt.scatter(lon, lat, c=np.log(population), cmap='viridis', s=area, linewidths=0, alpha=0.5)
plt.axis('equal')
plt.xlabel('longlatitude')
plt.ylabel('lattitude')
plt.colorbar(label='log$_{10}$(population)')
# plt.clim(4, 8)

area_range = [50, 100, 300, 500]
for a in area_range:
    plt.scatter([], [], s=a, c='k', alpha=0.4, label=str(a) + 'km$^2$')
    
plt.legend(labelspacing=1, scatterpoints=1, title='Cyty Area')
plt.title('California: Populatio and Area distribution')
```
